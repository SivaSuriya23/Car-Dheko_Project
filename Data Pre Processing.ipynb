```python
# Function to handle different formats
def convert_to_numeric(price_str):
    if isinstance(price_str, str):
        # Remove currency symbol and commas
        price_str = price_str.replace('₹', '').replace(',', '')
        
        # Determine multiplier based on suffix
        if 'Lakh' in price_str:
            price_str = price_str.replace(' Lakh', '').strip()
            multiplier = 100000
        elif 'Crore' in price_str:
            price_str = price_str.replace(' Crore', '').strip()
            multiplier = 10000000
        else:
            multiplier = 1
        
        # Convert to float and apply the multiplier
        try:
            return float(price_str) * multiplier
        except ValueError:
            return None

    return None

# Apply the function to the 'price' column
df['price'] = df['price'].apply(convert_to_numeric)
df['priceActual'] = df['priceActual'].apply(convert_to_numeric)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from scipy import stats

# 1. Standardize Data Formats
df['km'] = df['km'].str.replace(' kms', '')  # Remove ' kms'
df['km'] = df['km'].str.replace(',', '')  # Remove commas
df['km'] = df['km'].astype(float)  # Convert to float
```

```python
# Identify categorical columns (example columns; adjust as necessary)
nominal_columns = ['transmission', 'owner', 'oem', 'model', 'variantName', 'Fuel Type', 'Color', 'Engine Type']
ordinal_columns = ['Seats', 'No of Cylinder', 'Gear Box', 'Drive Type', 'Seating Capacity']

# 2. Encode Categorical Variables
# One-Hot Encoding for Nominal Categories
df = pd.get_dummies(df, columns=nominal_columns, drop_first=True)

# Label Encoding for Ordinal Categories
label_encoder = LabelEncoder()
for col in ordinal_columns:
    df[col] = label_encoder.fit_transform(df[col])

# Alternatively, for custom ordinal mapping:
# Example ordinal mapping for 'Seats' column (adjust values as necessary)
# seats_mapping = {'2': 1, '4': 2, '5': 3, '7': 4}
# df['Seats'] = df['Seats'].map(seats_mapping)
```

```python
# Drop columns NaN columns
df = df.drop(['priceSaving', 'priceTrendingText'], axis=1, errors='ignore')
```

```python

# Identify numerical columns that require cleaning (example columns; adjust as necessary)
numerical_columns_with_units = [
    'Engine Displacement',  # Example column with "cc" units
    # Add other similar columns that have units or text with numeric data
]

# Function to clean the numerical columns
def clean_numerical_column(column):
    # Ensure column is of string type
    column = column.astype(str)
    # Remove any non-numeric characters (like "cc" or "kms")
    column = column.str.replace(r'[^\d.]', '', regex=True)
    # Convert to float
    return pd.to_numeric(column, errors='coerce')

# Apply the cleaning function to the specified columns
df[numerical_columns_with_units] = df[numerical_columns_with_units].apply(clean_numerical_column)

# Add other numerical columns that don't need unit removal
numerical_columns = [
    'km', 'price', 'priceActual', 'Mileage', 
    'Max Power', 'Torque', 'Length', 'Width', 'Height', 
    'Wheel Base', 'Front Tread', 'Rear Tread', 'Kerb Weight', 'Gross Weight'
]

# Combine both cleaned and non-cleaned numerical columns
all_numerical_columns = numerical_columns + numerical_columns_with_units

# Check data types of all numerical columns to ensure they are float
for col in all_numerical_columns:
    if df[col].dtype != 'float64':
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Check for any remaining non-numeric values
print(df[all_numerical_columns].info())

# Drop rows with NaN values if needed
#df = df.dropna(subset=all_numerical_columns)

# 3. Normalize Numerical Features

# Option 1: Min-Max Scaling
min_max_scaler = MinMaxScaler()
df[all_numerical_columns] = min_max_scaler.fit_transform(df[all_numerical_columns])

# Option 2: Standard Scaling (uncomment if you want to use this instead)
#standard_scaler = StandardScaler()
#df[all_numerical_columns] = standard_scaler.fit_transform(df[all_numerical_columns])
```

```python
# IQR Method

# List of numerical columns
numerical_columns = [
    'km', 'price', 'priceActual', 'Mileage', 'Max Power', 'Torque', 
    'Length', 'Width', 'Height', 'Wheel Base', 'Front Tread', 
    'Rear Tread', 'Kerb Weight', 'Gross Weight'
]

# Remove columns not present in the DataFrame
numerical_columns = [col for col in numerical_columns if col in df.columns]

# Apply IQR method to each numerical column
for col in numerical_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    df = df[~((df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR)))]

print("DataFrame after IQR outlier removal:")
print(df)
```
