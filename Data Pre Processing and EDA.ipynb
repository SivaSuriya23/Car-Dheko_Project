```python
def clean_price(price):
    price = str(price).replace('₹', '').replace(',', '').strip()
    if 'Crore' in price:
        return float(price.replace('Crore', '').strip()) * 100  # Convert to lakhs
    elif 'Lakh' in price:
        return float(price.replace('Lakh', '').strip())  # Already in lakhs
    else:
        return float(price) / 100000  # Convert from other formats to lakhs

df['price'] = df['price'].apply(clean_price)
```

```python
import pandas as pd
import numpy as np
from scipy import stats

# Define numerical columns based on actual DataFrame columns
numerical_cols = ['Price', 'Seats', 'Mileage']

# Ensure the columns are numeric and handle NaN values
df[numerical_cols] = df[numerical_cols].apply(pd.to_numeric, errors='coerce')

# Check for NaN values
print("NaN values in each column:")
print(df[numerical_cols].isna().sum())

# Fill NaN values with column mean
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].mean())

# Calculate Z-scores
z_scores = np.abs(stats.zscore(df[numerical_cols]))

# Remove outliers
if z_scores.shape[0] == df.shape[0]:
    df_clean = df[(z_scores < 3).all(axis=1)]
    print("DataFrame shape after outlier removal:", df_clean.shape)
else:
    raise ValueError("Z-scores array shape does not match DataFrame shape.")
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor

# Define feature matrix (X) and target variable (y)
X = df_clean.drop('Original Equipment Manufacturer', axis=1)
y = df_clean['Price']

# Identify categorical columns
categorical_cols = X.select_dtypes(include=['object']).columns

# Define the preprocessor for categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ],
    remainder='passthrough'  # Keep non-categorical columns unchanged
)

# Create a pipeline that first preprocesses the data and then fits the model
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', RandomForestRegressor())
])

# Fit the pipeline
pipeline.fit(X, y)

# Get feature importances from the model
importances = pipeline.named_steps['model'].feature_importances_

# Extract feature names after one-hot encoding
one_hot_feature_names = pipeline.named_steps['preprocessor'].transformers_[0][1].get_feature_names_out(categorical_cols)
all_feature_names = list(one_hot_feature_names) + [col for col in X.columns if col not in categorical_cols]

# Create a DataFrame of feature importances
feature_importances = pd.DataFrame({'Feature': all_feature_names, 'Importance': importances})
print(feature_importances.sort_values(by='Importance', ascending=False))
```
